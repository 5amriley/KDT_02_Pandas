#### astype() 메서드를 사용하여 데이터 유형 변환

```python
# 다수의 열의 dtype 변경
# survived: bool, pclass: category, sex: category, embarked: category, class: category, 
# who: category, deck: category, embark_town: category
titanicDF = titanicDF.astype({'survived':'bool', 'pclass':'category', 'sex':'category', 'embarked':'category', 'class':'category',
                             'who':'category', 'deck':'category', 'embark_town':'category'})
titanicDF
```

```python
# DF의 각 열 별 결측치 확인
# 전체 891행 중 177행에서 age가 결측치 (-> age는 결측치를 채워준다.)
# 전체 891행 중 688행에서 deck이 결측치 (-> deck은 분석에 사용 X)
titanicDF.isna().sum()
```

```python
# age가 null인 행들을 필터링
no_age = titanicDF['age'].isnull()
titanicDF[no_age]
```

```python
# age가 null인 행들 중 어린이('child')는 없음 => 성인 남자 or 성인 여자
titanicDF[no_age]['who'].unique()
```

age가 NaN인 사람들 중에서 adult_male이 True인 경우도 있고, False인 경우도 있다. (나이, 성별에 대한 간접정보)  
성인남성에 해당한다면, 성인남성 나이의 평균치로 NaN을 대체하고,  
성인남성이 아니라면, 성인남성이 아닌 사람들의 평균치로 NaN을 대체할 것이다.

```python
# 나이 정보가 있는 성인
is_adult_male = titanicDF['adult_male'] == True
adult_male_with_age = titanicDF[(~no_age) & is_adult_male]
adult_male_with_age
```

```python
adult_male_with_age['age'].sum()
```

```python
adult_male_with_age['age'].count()
```

```python
# 나이 정보가 있는 성인들의 나이 평균
adult_male_avg_age = adult_male_with_age['age'].sum() / adult_male_with_age['age'].count()
adult_male_avg_age
```

```python
# adult_male==True인 행들의 age 결측치 대체
titanicDF.loc[no_age & is_adult_male, 'age'] = adult_male_avg_age
titanicDF[no_age & is_adult_male]
```

```python
# 나이 정보가 없는 사람들의 나이 평균
not_adult_male_with_age = titanicDF[(~no_age) & (~is_adult_male)]
not_adult_male_avg_age = not_adult_male_with_age['age'].sum() / not_adult_male_with_age['age'].count()
not_adult_male_avg_age
```

```python
# adult_male==False인 행들의 age 결측치 대체
titanicDF.loc[no_age & (~is_adult_male), 'age'] = not_adult_male_avg_age
titanicDF[no_age & (~is_adult_male)].head()
```

```python
# 생존률 분석 : 성별('sex')
is_survived = titanicDF['survived']   # bool Series가 반환된다. (titanicDF['survived']==True 할 필요 X)
print("'sex' 컬럼 분석")
resultDF = pd.DataFrame(columns = ['전체', '생존', '생존률(%)'])
for s in np.sort(titanicDF['sex'].unique()):
    the_condition = titanicDF['sex']==s
    total_col = len(titanicDF[the_condition])
    survived_col = len(titanicDF[the_condition & is_survived])
    #print(f'[{s}] 전체 : {total_col} 생존 : {survived_col}, 생존률(%) : {round(survived_col / total_col * 100, 2)}')
    resultDF.loc[s] = [total_col, survived_col, round(survived_col / total_col * 100, 1)]
resultDF
```

```python
# 생존률 분석 : 'sibsp'
is_survived = titanicDF['survived']   # bool Series가 반환된다. (titanicDF['survived']==True 할 필요 X)
print("'sibsp' 컬럼 분석")
resultDF = pd.DataFrame(columns = ['전체', '생존', '생존율(%)'])
for s in np.sort(titanicDF['sibsp'].unique()):
    the_condition = titanicDF['sibsp']==s
    total_col = len(titanicDF[the_condition])
    survived_col = len(titanicDF[the_condition & is_survived])
    #print(f'[{s}] 전체 : {total_col} 생존 : {survived_col}, 생존율(%) : {round(survived_col / total_col * 100, 2)}')
    resultDF.loc[s] = [total_col, survived_col, round(survived_col / total_col * 100, 1)]
resultDF
```

```python
# 생존률 분석 : 'age'
is_survived = titanicDF['survived']
buckets = [titanicDF['age'].min(), 5, 12, 18, 29, 49, 64, titanicDF['age'].max()]
survived_age_buckets = titanicDF[is_survived]['age'].value_counts(bins=buckets)
survived_age_buckets.sort_index()   # Series
```

```python
total_age_buckets = titanicDF['age'].value_counts(bins=buckets)
total_age_buckets.sort_index()   # Series
```

```python
age_survival_analysis = round(survived_age_buckets / total_age_buckets * 100, 2)
#age_survival_analysis = age_survival_analysis.to_frame()   # Series를 DataFrame으로 변환
#age_survival_analysis.columns = ['생존율(%)']
#age_survival_analysis.index = ['0~5세', '6~12세', '13~18세', '19~29세', '30~49세', '50~64세', '65세 이상']
age_survival_analysis
```

```python
# 각 컬럼별 생존율 분석 함수
import math
def survival_rate_analysis(col_name: str):
    # titanicDF 전용 함수 (titanicDF가 전역변수로 생성되어 있어야 함)
    if col_name not in titanicDF.columns:
        print('존재하지 않는 열입니다.')
        return
    elif col_name == 'survived':
        print('다른 열을 입력해주세요.')
        return

    is_survived = titanicDF['survived']   # bool Series가 반환된다. (titanicDF['survived']==True 할 필요 X)
    print(f"'{col_name}' 컬럼 분석")
    
    if titanicDF[col_name].nunique() > 10:
        # 최대, 최소값이 어떤 값이 되든 7구간으로 나눠지도록 buckets을 설정하는 알고리즘? (마지막 구간의 길이는 달라도 됨, 단, 모든 구간의 길이는 정수여야 함)
        #buckets = [titanicDF['age'].min(), 5, 12, 18, 29, 49, 64, titanicDF['age'].max()]
        buckets = np.arange(round(titanicDF[col_name].min()), round(titanicDF[col_name].max()), 
                            round((titanicDF[col_name].max() - titanicDF[col_name].min()) / 8))
        #print(buckets) # 디버그용
        buckets[-1] = math.ceil(titanicDF[col_name].max())
        #print(buckets) # 디버그용
        survived_age_buckets = titanicDF[is_survived][col_name].value_counts(bins=buckets)
        total_age_buckets = titanicDF[col_name].value_counts(bins=buckets)
        age_survival_analysis = round(survived_age_buckets / total_age_buckets * 100, 1)
        # 여러 Series를 하나의 DataFrame으로 생성 (열이름도 지정)
        resultDF = pd.DataFrame({'전체':total_age_buckets, 
                                 '생존':survived_age_buckets, 
                                 '생존율(%)':age_survival_analysis}).sort_index()
    else: 
        resultDF = pd.DataFrame(columns = ['전체', '생존', '생존율(%)'])   # 열 정보만 있는, 빈 DataFrame 생성
        for s in np.sort(titanicDF[col_name].unique()):
            the_condition = titanicDF[col_name]==s
            total_col = len(titanicDF[the_condition])
            survived_col = len(titanicDF[the_condition & is_survived])
            # DataFrame에 행을 하나씩 추가 (total_col, survived_col 모두 int인데, round()한 값이 float라 모두 float로 저장된다.)
            resultDF.loc[s] = [total_col, survived_col, round(survived_col / total_col * 100, 1)]
        resultDF = resultDF.astype({'전체':'int32', '생존':'int32'})  # 정수값이 저장되는 '전체', '생존' 열은 int로 형변환
    return resultDF
```
