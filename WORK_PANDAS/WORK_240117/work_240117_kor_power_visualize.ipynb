```python
import pandas as pd
```

```python
electricity_file = './남북한발전전력량.xlsx'
korea_elec = pd.read_excel(electricity_file, index_col=[0, 1])
korea_elec
```

#### 결측치 대체 ( -을 0로 )

```python
korea_elec.loc[('남한', '신재생')] = korea_elec.loc[('남한', '신재생')].replace('-', '0')
korea_elec.loc[('북한', '원자력')] = korea_elec.loc[('북한', '원자력')].replace('-', '0')
korea_elec
```

```python
# DataFrame 전체 값을 int로 형변환
korea_elec = korea_elec.astype('int')
```

```python
korea_elec.info()
```

# 전력량 시각화 분석

```python
import matplotlib.pyplot as plt          # 그래프 관련 모듈
import matplotlib as mpl                 # 패키지 관련
import matplotlib.font_manager as fm     # 패키지 리소스 중 폰트 관련 모듈
```

- 한글 폰트 설정

```python
# 폰트 파일의 경로 저장
font_path = r'C:\Windows\Fonts\gulim.ttc'
```

```python
# matplotlib 패키지에 한글 폰트 설정
myFont = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=myFont)
```

```python
color_list = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
```

- [남한] 수력 vs. 화력

```python
# 그래프 그리기
plt.plot(korea_elec.columns, korea_elec.loc[('남한', '수력')], 'b', korea_elec.columns, korea_elec.loc[('남한', '화력')], 'r')
plt.title('[남한] '+'수력' + ' vs. ' + '[남한] ' + '화력')

# x축, y축 레이블
plt.xlabel('년도', loc='right', fontsize='large')
plt.ylabel('발전 전력량', loc='top', rotation='horizontal')

# x축, y축 눈금 설정
# x축 값으로 설정된 korea_elec.columns 값들이 모두 object이므로, tick을 문자열로 해야 한다.
year_ticks = list(map(str, range(1990, 2017, 1)))
year_labels = [s if i % 3 == 0 else '' for i, s in enumerate(year_ticks)]
plt.xticks(ticks=year_ticks, labels=year_labels)
plt.yticks(color='blue')

# 범례
plt.legend(['[남한] 수력', '[남한] 화력'])

plt.grid(ls='--')
plt.show()
```

- [남한] 수력, 화력, 원자력, 신재생 비교

```python
# 그래프 그리기
plt.plot(korea_elec.columns, korea_elec.loc[('남한', '수력')], 'b', korea_elec.columns, korea_elec.loc[('남한', '화력')], 'r',
        korea_elec.columns, korea_elec.loc[('남한', '원자력')], 'y', korea_elec.columns, korea_elec.loc[('남한', '신재생')], 'g')
plt.title('[남한] '+'수력' + ' vs. ' + '[남한] ' + '화력' + ' vs. ' + '[남한] ' + '원자력' + ' vs. ' + '[남한] ' + '신재생')

# x축, y축 레이블
plt.xlabel('년도', loc='right', fontsize='large')
plt.ylabel('발전 전력량', loc='top', rotation='horizontal')

# x축, y축 눈금 설정
# x축 값으로 설정된 korea_elec.columns 값들이 모두 object이므로, tick을 문자열로 해야 한다.
year_ticks = list(map(str, range(1990, 2017, 1)))
year_labels = [s if i % 3 == 0 else '' for i, s in enumerate(year_ticks)]
plt.xticks(ticks=year_ticks, labels=year_labels)
plt.yticks(color='blue')

# 범례
plt.legend(['[남한] 수력', '[남한] 화력', '[남한] 원자력', '[남한] 신재생'])

plt.grid(ls='--')
plt.show()
```
