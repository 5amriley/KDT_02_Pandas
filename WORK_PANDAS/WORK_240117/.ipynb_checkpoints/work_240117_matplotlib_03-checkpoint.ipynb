# 3. 주석 사용

- 제목 추가

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X- 2.)

# 제목 추가
plt.title('A polynomial')
plt.plot(X, Y, c = 'b')   # c (color) 속성 설정
plt.show()
```

- 라텍스 스타일 표기법 사용

```python
plt.title('$f(x)=\\frac{1}{4}(x+4)(x+1)(x-2)$')
plt.plot(X, Y, c = 'k')
plt.show()
```

- 각 축에 레이블 추가

```python
plt.title('Power curve for airfoil KV873')
plt.xlabel('Air speed')    # x축 레이블
plt.ylabel('Total drag')   # y축 레이블
plt.plot(X, Y, c = 'k')
plt.show()
```

- 텍스트 추가

```python
plt.text(-0.5, -0.25, 'Brackmard minimum')   # 텍스트 추가

plt.plot(X, Y, c = 'k')
plt.show()
```

```python
box = {
    'facecolor': '.75',
    'edgecolor': 'k',
    'boxstyle': 'round'
}

plt.text(-0.5, -0.20, 'Brackmard minimum', bbox=box)

plt.plot(X, Y, c='k')
plt.show()
```

- 화살표 추가

```python
plt.annotate('Brackmard minimum',
            ha='center',  # horizontalalignment
            va='bottom',  # verticalalignment
            xytext=(-1.5, 3.),
            xy=(0.75, -2.7),
            arrowprops={'facecolor':'black', 'shrink':0.05})

plt.plot(X, Y)
plt.show()
```

- 범례 추가

```python
X = np.linspace(0, 6, 1024)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.xlabel('X')
plt.ylabel('Y')

plt.plot(X, Y1, c='k', lw=3., label='sin(X)')            # 곡선 추가 (+레이블 설정)
plt.plot(X, Y2, c='.5', lw=3., ls='--', label='cos(X)')  # 곡선 추가 (+레이블 설정)

plt.legend()   # 범례 추가
plt.show()
```

- 격차 추가

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.plot(X, Y, c='k')   # 곡선 추가
plt.grid(True, lw=2, ls='--', c='.75')   # 격차 추가
plt.show()
```
