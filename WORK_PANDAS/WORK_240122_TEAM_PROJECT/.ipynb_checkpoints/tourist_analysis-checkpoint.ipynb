# 한국인 일본여행 통계 (2004~2023)

```python
import pandas as pd
```

```python
travel_file = './국민 해외관광객 주요 목적지별 통계(202311).xlsx'
```

```python
travelDF = pd.read_excel(travel_file, sheet_name='Asia', index_col=[0, 1], header=[1, 2])
travelDF = travelDF[[('법무부・KTO','명수'), ('법무부・KTO','전년대비'), ('일본','명수'), ('Japan', '전년대비')]]
travelDF
```

```python
# 2000~2003년의 경우, 월별 여행객 자료는 없지만 한 해 누계 자료는 존재.
travelDF[-25:-1]
```

```python
# 연간 여행객, 월별 여행객 자료 분리
yearly_travelDF = travelDF[-25:-1]
travelDF = travelDF[:-25]
travelDF   # 월별
```

- 한글 폰트 설정

```python
import matplotlib as mpl                 # 패키지 관련
import matplotlib.font_manager as fm     # 패키지 리소스 중 폰트 관련 모듈
```

```python
# 폰트 파일의 경로 저장
font_path = r'C:\Windows\Fonts\gulim.ttc'
```

```python
# matplotlib 패키지에 한글 폰트 설정
myFont = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=myFont)
```

```python
travelDF.index
```

```python
# travelDF의 인덱스를 datetime 객체로 만들기 위해 형식 변경하여 저장
dateList = ['-'.join([y[:-1], m[:-1], '1']) for y, m in travelDF.index]
dateList
```

```python
# travelDF의 인덱스를 datetime 객체로 바꾼 것 => dateDF
dateDF = pd.DataFrame(dateList, columns=['date'])
dateDF = dateDF.astype('datetime64[ns]')
dateDF.info()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(dateDF.date[:-1], travelDF[('법무부・KTO','명수')].values[:-1], color='g', label='전체 여행객')
ax.plot(dateDF.date[:-1], travelDF[('일본','명수')].values[:-1], color='r',  label='일본 여행객')
dateFmt = mdates.DateFormatter('%Y-%m')
ax.xaxis.set_major_formatter(dateFmt)
plt.xticks()
plt.ylabel('\n여행객 수\n(백만명)', loc='top', rotation='horizontal')
plt.grid()
plt.legend(fontsize=11)
plt.show()
#plt.xticks([t for t in dateList if int(t[:4])%4==0 and t[4:7]=='-1-'], rotation=45)
```

```python
the_years = (dateDF['date'].dt.year == 2019) | (dateDF['date'].dt.year == 2020)
```

```python
the_years
```

```python
dateDF2 = dateDF[the_years]
dateDF2
```

```python
# travelDF 인덱스 초기화 => the_years 사용가능 (숫자 인덱스로 동일해지므로)
travelDF2 = travelDF.reset_index(drop=True)[the_years]
travelDF2
```

```python
fig, ax = plt.subplots(figsize=(15, 7))
ax.plot(dateDF2.date, travelDF2[('법무부・KTO','명수')].values, color='g', marker = 'o', markersize = 5, label='전체 여행객')
ax.plot(dateDF2.date, travelDF2[('일본','명수')].values, color='r', marker = 'o', markersize = 5, label='일본 여행객')
dateFmt = mdates.DateFormatter('%Y-%m')
ax.xaxis.set_major_formatter(dateFmt)
plt.xticks()
plt.ylabel('\n여행객 수\n(백만명)', loc='top', rotation='horizontal')
plt.grid()
plt.legend(fontsize=11)
plt.show()
```
