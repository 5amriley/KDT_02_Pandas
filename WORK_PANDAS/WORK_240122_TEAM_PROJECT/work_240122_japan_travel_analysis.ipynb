# 한국인 일본여행 통계 (2004~2023)

```python
import pandas as pd
```

```python
travel_file = './국민 해외관광객 주요 목적지별 통계(202311).xlsx'
```

```python
travelDF = pd.read_excel(travel_file, sheet_name='Asia', index_col=[0, 1], header=[1, 2])
travelDF = travelDF[[('법무부・KTO','명수'), ('법무부・KTO','전년대비'), ('일본','명수'), ('Japan', '전년대비')]]
travelDF
```

```python
# 2000~2003년의 경우, 월별 여행객 자료는 없지만 한 해 누계 자료는 존재.
travelDF[-25:-1]
```

```python
# 연간 여행객, 월별 여행객 자료 분리
yearly_travelDF = travelDF[-25:-1]
travelDF = travelDF[:-25]
travelDF   # 월별
```

```python
travelDF.index
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 7))
plt.plot([y+m for y, m in travelDF.index], travelDF[('일본','명수')].values, 'g')
plt.xticks([y+m for y, m in travelDF.index if m=='1월' and int(y[:-1]) % 2 == 0], 
           [y[:-1] for y, m in travelDF.index if m=='1월' and int(y[:-1]) % 2 == 0])
plt.axvline('2019년7월', color='r')
plt.grid()
plt.show()
```

```python
import numpy as np
X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X- 2.)

# 제목 추가
plt.title('A polynomial')
plt.plot(X, Y, c = 'b')   # c (color) 속성 설정
plt.show()
```

```python
import matplotlib.dates as mdates

date = ['1975-12-1', '1988-8-15', '1998-1-17', '2002-6-30', '2008-08-20', '2011-3-16']
value = [1,4,5,9,10,12]
df = pd.DataFrame({
    'date': date,
    'value': value
})

df['date'] = pd.to_datetime(df['date'])
df
```

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(df['date'], df['value'], marker='o', markersize=10)
dateFmt = mdates.DateFormatter('%Y-%m')
ax.xaxis.set_major_formatter(dateFmt)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
