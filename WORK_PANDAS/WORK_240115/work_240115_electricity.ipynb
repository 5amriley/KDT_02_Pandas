```python
import pandas as pd
```

#### (1) 파일 읽어오기

```python
electricity_file = './남북한발전전력량.xlsx'
korea_elec = pd.read_excel(electricity_file, index_col=[0, 1])
korea_elec
```

```python
korea_elec.index
```

#### 결측치 대체 ( -을 0로 )

```python
korea_elec.loc[('남한', '신재생')] = korea_elec.loc[('남한', '신재생')].replace('-', '0')
korea_elec.loc[('북한', '원자력')] = korea_elec.loc[('북한', '원자력')].replace('-', '0')
korea_elec
```

#### 북한 신재생 행 추가

```python
korea_elec.loc[('북한', '신재생'),:] = 0
```

```python
korea_elec
```

```python
korea_elec.info()
```

#### 데이터 타입 변환 (object -> int32)

```python
for col in korea_elec.columns:
    korea_elec[col] = korea_elec[col].astype(int)
korea_elec.info()
```

#### 마지막 열 이후에 총계 추가

```python
korea_elec.columns
```

```python
sum_col = []
for r in korea_elec.index:
    sum_col.append(korea_elec.loc[r].sum())
sum_col
```

```python
korea_elec['합계'] = sum_col
korea_elec
```
