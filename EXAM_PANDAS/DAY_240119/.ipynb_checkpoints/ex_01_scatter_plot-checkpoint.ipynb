```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([(.2, .3), (.0, .6), (.6, .0), (.2, .1)],
                columns=['dogs', 'cats'])
df
```

```python
def histogram_intersection(a, b):
    # np.minimum() => 같은 차원의 array들을 입력으로 받음. 각 위치 별 최소값을 구하여 결과를 반환.
    v = np.minimum(a, b).sum().round(decimals=1)
     return v
df.corr(method=histogram_intersection)
```

```python
df.corr()    # 피어슨 방식 (두 변수간의 선형 상관관계. -1 ~ +1)
```

```python
df1 = pd.DataFrame([(1, 1), (2, np.nan), (np.nan, 3), (4, 4)],
                  columns=['dogs', 'cats'])
df1
```

```python
df1.corr()
```

```python
df1.corr(numeric_only=True, min_periods=2)
```

```python
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3, include_lowest=True)
```

### 시각화로 연비 mpg 컬럼의 관계 확인 ----------------------

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(mpgDF['mpg'], mpgDF['weight'])
plt.title(f'mpg - weight [{corrDF["mpg"]["weight"]}]')
plt.xlabel('mpg')
plt.ylabel('weight')
plt.show()
```

```python
def scatterplot(sr1, sr2):
    # parameters :
    # - sr1, sr2 : series instance
    plt.scatter(sr1, sr2)
    plt.title(f'{sr1.name} - {sr2.name} [{round(corrDF[sr1.name][sr2.name], 2)}]')
    plt.xlabel(sr1.name)
    plt.ylabel(sr2.name)
    plt.show()
```

```python
def printScatter(col1, col2):
    # parameters :
    # - col1, col2 : column name
    plt.scatter(mpgDF[col1], mpgDF[col2])
    plt.title(f'[{col1} - {col2}]')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
```

```python
for col in corrDF.columns[1:]:
    printScatter('mpg', col)
```

- 다중 그래프

```python
plt.figure(figsize=(15, 10))   # 인치(inch) 단위

plt.subplot(2, 2, 1)
plt.scatter(mpgDF['mpg'], mpgDF['displacement'])
plt.title('mpg-displacement')

plt.subplot(2, 2, 2)
plt.scatter(mpgDF['mpg'], mpgDF['weight'])
plt.title('mpg-weight')

plt.subplot(4, 1, 3)
plt.scatter(mpgDF['mpg'], mpgDF['cylinders'])
plt.title('mpg-cylinders')

plt.tight_layout()   # 여러 개의 그래프가 겹치지 않게 최소한의 여백 지정해주는 함수
plt.show()
```

```python
fig, ax = plt.subplots()   # fig는 테두리

ax1 = fig.add_subplot(3, 1, 1)
ax1.scatter(mpgDF['mpg'], mpgDF['displacement'])
ax1.set_title('mpg-displacement')

ax2 = fig.add_subplot(3, 1, 2)
ax2.scatter(mpgDF['mpg'], mpgDF['weight'])
ax2.set_title('mpg-weight')

ax3 = fig.add_subplot(3, 1, 3)
ax3.scatter(mpgDF['mpg'], mpgDF['cylinders'])
ax3.set_title('mpg-cylinders')

plt.tight_layout()   # 여러 개의 그래프가 겹치지 않게 최소한의 여백 지정해주는 함수
plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)   # x = [0, 2*pi] 구간을 400 points로 분리
y = np.sin(x**2)                   # y = sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)   # y축을 공유 (y축 눈금이 한 군데만 표시)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
# Create four polar axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
```

```python
# Share a X axis with each column of subplots
plt.subplots(2, 2, sharex='col')

# Share a Y axis with each row of subplots
plt.subplots(2, 2, sharey='row')

# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')

# Note that this is the same as
plt.subplots(2, 2, sharex=True, sharey=True)

# Create figure number 10 with a single subplot
# and clears it if it already exists.
fig, ax = plt.subplots(num=10, clear=True)
```
