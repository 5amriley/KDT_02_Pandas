### 최빈값, 중앙값, 평균 을 통해 분포를 알아볼 수 있다. (좌우대칭? 한쪽 쏠림?)

- 컬럼과 컬럼 사이의 관계를 수치화 해주는 메서드 : df.corr()
- 범위 : -1 ~ 1 (-1에 가까울 수록, 1에 가까울 수록 상관관계가 높은 것)

```python
mpgDF.corr(numeric_only=True)
```

* 컬럼명 지정

```python
mpgDF.columns = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"]
```

```python
mpgDF
```

```python
mpgDF.corr(numeric_only=True)
```

- 데이터타입과 실제 데이터가 다른 컬럼 ==> 형변환

```python
# '?'를 어떻게 처리할 것인가?
```

```python
mpgDF.horsepower.value_counts()['?']
```

```python
# copy() 없이 실행할 경우 Warning이 뜰 수 있다.
horseSR = mpgDF.horsepower.copy()
```

```python
for idx, horse in enumerate(horseSR):
    if horse == '?':
        horseSR[idx] = '0.0'
```

```python
horseSR.unique()
```

```python
mpgDF.horsepower = horseSR.astype('float64')
```

```python
mpgDF.info()
```

```python
mpgDF.corr(numeric_only=True).mpg.sort_values()
```

```python
# origin 컬럼의 타입을 int64 -> category로 변환
# 왜? 1-USA, 2-JP, 3-EU 이기 때문에 수치적인 의미가 없다. (상관관계 분석 시 무의미)
mpgDF.origin = mpgDF.origin.astype('category')
```

```python
mpgDF.info()
```

# [데이터 원소값 변경 / 타입 변경 한꺼번에 진행 ]

```python
# 데이터 읽어와서 바로 컬럼명 지정
mpgDF2 = pd.read_csv(file, header=None, sep='\s+', names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"])
```

```python
# 데이터 정보 확인
mpgDF2.info()
```

```python
# 원소값 '?' 변경해주는 메서드 => replace(old, new, inplace)
mpgDF2.horsepower.replace('?', '0.0', inplace=True)
```

```python
# origin 컬럼의 1- USA, 2- EU, 3- JPN
mpgDF2.origin.replace({1:'USA', 2:'EU', 3:'JPN'}, inplace=True)
```

```python
# origin 컬럼의 값의 종류별 개수 확인 => value_counts()
mpgDF2.origin.value_counts()
```

```python
# origin 컬럼의 값의 종류 확인 => unique()
mpgDF2.origin.unique()
```

```python
# 고유값 즉 값의 종류와 개수
mpgDF2.origin.nunique()
```

- horsepower 컬럼과 origin 컬럼의 데이터 타입 변경

```python
mpgDF2.info()
```

```python
# mpgDF2.horsepower.astype('int64') # 왜 안 되는 걸까
mpgDF2.horsepower = mpgDF2.horsepower.astype('float64')
```

```python
mpgDF2.origin = mpgDF2.origin.astype('category')
```

```python
mpgDF2.info()
```
