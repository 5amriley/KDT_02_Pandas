<hr> (1) 모듈 로딩 <hr>

```python
import pandas as pd
```

<hr> (2) 데이터 준비 <hr>

```python
datafile = "../../DATA/movies.csv"
```

<hr> 
(3) 데이터 저장 => DataFrame  

* 파일 데이터 읽어서 저장 => pandas.read_파일확장자()
<hr>

```python
movieDF = pd.read_csv(datafile)
```

```python
movieDF
```

- 데이터 확인 관련 메서드

```python
# 데이터의 전체적인 구조 및 데이터 종류 요약
movieDF.info()
```

```python
# 컬럼별 통계적인 데이터 정보 제공
twoDF = movieDF.describe()
twoDF
```

```python
# 타입에 상관없이 모든 데이터 정보 제공
movieDF.describe(include='all')
```

```python
# 수치 컬럼의 '최소, 최대, 평균' 데이터 추출
twoDF.loc[['min', 'max', 'mean']]
```

- 행 -> 열, 열 -> 행 치환 기능 속성 및 메서드

```python
pd.DataFrame([twoDF.Rank.values, twoDF.Year.values], columns=twoDF.index, index=twoDF.columns)
```

- 속성 : DF객체변수명.T  => 원본 데이터 변경 안 됨! (비파괴적)

```python
twoDF.T
```

- 메서드 : 

```python
twoDF.transpose()
```

<hr> (5) 데이터 정렬  

 * 인덱스 기반 정렬 => 행인덱스 / 열인덱스
<hr>

```python
# 행 인덱스 기반 정렬  => DF.sort_index()
# axis가 0:행
movieDF.sort_index(ascending=False)
```

```python
# 컬럼 인덱스 기반 정렬 => 내림차순
movieDF.sort_index(axis='columns', ascending=False)
```

<hr> (5) 데이터 정렬  

 * 데이터/값 기반 정렬
<hr>

```python
movieDF.sort_values(by='Year')
```

```python
movieDF.sort_values(by=['Year', 'Title'])
```

- [실습] 다양한 정렬

```python
from math import nan

data={'col1':['A','A','B',nan,'D','C'],
      'col2':[2,1,9,8,7,4],
      'col3':[0,1,9,4,2,3],
      'col4':['a','B','c','D','e','F']}
```

```python
dataDF = pd.DataFrame(data)
dataDF
```

- col 을 기준으로 정렬

```python
# NaN(빈칸 데이터) 값의 위치 설정: (기본) 제일 마지막 'last'
dataDF.sort_values(by='col1')
```

```python
# NaN(빈칸 데이터) 값의 위치 : na_position 파라미터
dataDF.sort_values(by='col1', na_position=)
```

```python
# col4 컬럼을 기준으로 오름차순 정렬
dataDF.sort_values(by='col4')
```

```python
dataDF.sort_values(by='col4', key=lambda col:col.str.upper())
```
