## [ 데이터 시각화 다듬기 ]
- x, y 축 레이블 지정
- x, y 축 눈금 설정
- 한글 폰트 설정
- 범례 설정

(1) 모듈 로딩

```python
import matplotlib.pyplot as plt          # 그래프 관련 모듈
import matplotlib as mpl                 # 패키지 관련
import matplotlib.font_manager as fm     # 패키지 리소스 중 폰트 관련 모듈
import pandas as pd
```

(2) 데이터 준비

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = range(1, 100, 10)
y = [d**2 for d in x]
```

(3) 그래프 그리기  
    * 시간의 변화를 표현해주는 그래프  
    * 연속석/순서 있는 데이터를 시각화할 경우 주로 사용됨

- [기본] 선 그래프

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축에 data1, y축에 data2 데이터로 사용
# 반환값 : 라인/선 그래프가 객체 리스트에 담겨서 반환됨.
ret = plt.plot(x, y)

# 퍼아썬에서 내장 출력 기능 함수 => print() 와 기능 동일
plt.show()
```

```python
# plt는 리스트를 반환. 그 리스트의 0번째 인덱스는 .Line2D 객체이다.
type(ret), len(ret), type(ret[0])
```

```python
ret[0].get_xdata()
```

- 라인/선 그래프에서 형식 즉 포맷지언 => str 타입의 포맷 설정
    * '색상형태선스타입' => 'ro--' : red circle odt
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b = blue, k = black
    * 대표적으로 많이 사용되는 형태 : o - 동그라미, ^ - 위로 삼걱형 < - 왼쪽 삼각형, ∨ - 아래로 삼각형
    * 대표적으로 많이 사용되는 선 스타일 : : - 실선, -- 점선, -. : dashdot(점과 직선이 번갈아)

```python
ret = plt.plot(x, y, 'ro')
```

```python
ret = plt.plot(x, y, 'ro--')
```

```python
ret = plt.plot(x, y, 'g^--')
```

## 그래프 제목 설정
    * 한글 폰트 미지원 => 한글 폰트 설정
    * 그래프에 제목 설정

- 한글 폰트 설정

```python
# 폰트 파일의 경로 저장
font_path = r'C:\Windows\Fonts\gulim.ttc'
```

```python
# matplotlib 패키지에 한글 폰트 설정
myFont = fm.FontProperties(fname=font_path).get_name()
mpl.rc('font', family=myFont)
```

- 그래프에 제목 설정

```python
x2 = [d // 2 for d in x]
y2 = [d // 2 for d in y]
plotList = plt.plot(x, y, 'ro--', x2, y2, 'g^-.')
plt.title("[테스트]")
```

```python
print(len(plotList))
```

### X, Y 축 레이블 설정 및 눈금 설정
- X, Y 축 데이터에 대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정  => plt.xticks(), plt.yticks(), plt.tick_param()

```python
# 그래프 그리기
plotList = plt.plot(x, y, 'ro--', x2, y2, 'g^-.')
plt.title("[테스트]")

# x축, y축 레이블
plt.xlabel('x축', loc='right', fontsize='large')
plt.ylabel('y축', loc='top', rotation='horizontal')
# 사용법 참고:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text  <- **kwargs 에 해당 (Text)

# x축, y축 눈금 설정
plt.xticks(ticks=[0, 2, 4, 6, 8], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')
```

### 범례(legend) 설정
- 여러 개의 그래프를 표시하는 경우 그래프별 식별을 위해서 사용

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^-.', label='그냥 결과')
plt.title("[테스트]")

# x축, y축 레이블
plt.xlabel('x축', loc='right', fontsize='large')
plt.ylabel('y축', loc='top', rotation='horizontal')
# 사용법 참고:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text  <- **kwargs 에 해당 (Text)

# x축, y축 눈금 설정
plt.xticks(ticks=[0, 2, 4, 6, 8], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')

# 범례
plt.legend(['red 라벨', 'green 라벨'])
# 사용법 참고:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend

# 그래프 안에 텍스트 출력
plt.text(0, 2000, "Happy New Year")
```

### 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y, 출력텍스트)

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^-.', label='그냥 결과')
plt.title("[테스트]")

# x축, y축 레이블
plt.xlabel('x축', loc='right', fontsize='large')
plt.ylabel('y축', loc='top', rotation='horizontal')
# 사용법 참고:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
# https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text  <- **kwargs 에 해당 (Text)

# x축, y축 눈금 설정
plt.xticks(ticks=[0, 2, 4, 6, 8], labels=['A', 'B', 'C', 'D', 'E'])
plt.yticks(color='blue')

# 범례
plt.legend(['red 라벨', 'green 라벨'])
# 사용법 참고:
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html#matplotlib.pyplot.legend

# 그래프 안에 텍스트 출력
plt.text(0, 2000, "Happy New Year", backgroundcolor='yellow', color='blue')

box={'facecolor':'0.75', 'edgecolor':'k', 'boxstyle':'round'}
plt.text(2, 4000, "Happy New Year", bbox=box)
```
