## [데이터 기반 시각화]
- 데이터 : 남북한발전전력.xlsx
- 남북한 수력발전전력량에 대한 시각화

(1) 모듈 로딩

```python
import pandas as pd
import matplotlib.pyplot as plt
```

(2) 데이터 준비

```python
file = '../../WORK_PANDAS/WORK_240115/남북한발전전력량.xlsx'
```

- 실제 데이터 파일 확인 결과
* 첫 번째 행은 컬럼명 데이터
* 첫 번째와 두 번째 컬럼 : 행 인덱스 사용에 적합
* 입력값이 없는 셀에 '-' 채워져 있음 => isnull()/isna()로는 결측값 체크 X
    - 삭제하지 않고 대체값 적용 => 대체값 0

(3) 데이터 저장 : Excel -> DataFrame

```python
korpowerDF = pd.read_excel(file, index_col=[0, 1])
```

(4) 데이터 확인

(4-1) 데이터 타입과 실제 데이터 체크  
    * '-' 데이터를 '0'으로 변경  
    * obejct 타입을 받아 uint16로 변경

```python
korpowerDF
```

```python
korpowerDF.info()
```

```python
print(korpowerDF.index, korpowerDF.columns, sep='\n\n')
```

(5) 데이터 전처리

(5-1) 데이터 자료형 변환

```python
# '-' -> '0'
korpowerDF2 = korpowerDF.replace('-','0')
```

```python
# object -> uint16
korpowerDF2 = korpowerDF2.astype('int16')
korpowerDF2.info()
```

(6) 데이터 분석 작업

- 남북한의 수력발전량 비교하기

```python
# 수력발전량 행 데이터 추출
waterDF = korpowerDF2.xs('수력', level=1)
waterDF
```

```python
plt.plot(waterDF.columns, waterDF.loc['남한'], waterDF.columns, waterDF.loc['북한'])
```

```python
plt.plot(waterDF.loc['남한'], 'bo-', waterDF.loc['북한'], 'ro-')
plt.show()
```
