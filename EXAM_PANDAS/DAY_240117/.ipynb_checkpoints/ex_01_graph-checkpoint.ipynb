## [ 데이터 시각화 맛보기 ]

(1) 모듈 로딩

```python
import matplotlib.pyplot as plt
import pandas as pd
```

(2) 데이터 준비

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = range(1, 100, 10)
y = [d**2 for d in x]
```

(3) 그래프 그리기

```python
ret = plt.plot(x, y)
```

```python
# plt는 리스트를 반환. 그 리스트의 0번째 인덱스는 .Line2D 객체이다.
type(ret), len(ret), type(ret[0])
```

```python
ret[0].get_xdata()
```

```python
ret = plt.plot(x, y, 'ro')
```

```python
ret = plt.plot(x, y, 'ro--')
```

```python
ret = plt.plot(x, y, 'g^--')
```

```python
x2 = [d // 2 for d in x]
y2 = [d // 2 for d in y]
plotList = plt.plot(x, y, 'ro--', x2, y2, 'g^--')
```
