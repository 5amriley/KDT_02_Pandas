```python
ord('a')
```

```python
sevens = pd.Series(sevens_data, index=sevens_index, dtype='byte')
sevens
```

[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
stock_file = '../../DATA/stock price.xlsx'
```

```python
stock = pd.read_excel('../../DATA/stock price.xlsx', index_col='id')
stock
```

3. 결측치 처리 (가)

```python
stock1 = stock.dropna()
stock1
```

```python
value_below_10000 = stock1['value'] <= 10000
stock1 = stock1[~value_below_10000]
```

3. 결측치 처리 (나)

```python
# 평균값으로 결측치(NaN) 대체
price_avg = stock['price'].sum() / stock['price'].notnull().sum()
stock.fillna(price_avg)
```

```python
# 이전 값으로 결측치(NaN) 대체
stock.fillna(method='ffill')
```

```python
# 이후 값으로 결측치(NaN) 대체
stock.fillna(method='bfill')
```

```python
value_1000 = stock1['value'] <= 1000
price_90000 = stock1['price'] >= 90000
stock1[value_1000 | price_90000]
```
